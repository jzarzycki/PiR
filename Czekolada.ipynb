### Powyższe zmienne można z łatwością zamienić na zmienne logiczne

````python
import re

cols = [
    "cocoa_butter", "vanilla",
    "lecithin", "salt", "sugar",
    "sweetener_without_sugar"
]

regex = re.compile("^have_not_")
str_to_bool = lambda s: not bool(regex.match(s))

for col in cols:
    choco[col] = choco[col].map(str_to_bool)

choco.info()
````

## Korelacja
### Cukier, oraz słodzik wykazują dużą korelację ujemną.
Jest to zrozumiałe, ponieważ słodzik jest zwykle zamiennikiem dla cukru. Postanowiłem nie usuwać na razie żadnej z tych zmiennych, ponieważ interesuje mnie, czy jest duża różnica w ocenach zbieranych przez czekoladę słodzoną słodzikiem.
### Liczba składników koreluje dosyć mocno z wanilią, lecytyną, oraz masłem kakaowym
Na stronie, z której pobrałem zestaw danych nie zostało przedstawione dokładne wyjaśnienie, jakie składniki obejmuje "liczba składników", co w połączeniu ze sporą korelacją tej zmiennej z innymi zmiennymi sprawia, że podejrzewam, że wliczane w nią mogą być właśnie te składniki. W związku z tym zdecydowałem się usunąć tą zmienną.

````python
choco.drop("counts_of_ingredients", axis=1, inplace=True)
choco.head()
````

````python
y = choco["rating"]
y.head()
````

````python
X = choco.drop("rating", axis=1)
X.head()
````

````python
X_ucz, X_test, y_ucz, y_test = sklearn.model_selection.train_test_split(
    X, y, test_size=0.2, random_state=9063
)
print(X_ucz.shape)
print(X_test.shape)
print(y_ucz.shape)
print(y_test.shape)
````

````python
import sklearn.linear_model
mnk = sklearn.linear_model.LinearRegression()
````

````python
def fit_regression(X_ucz, X_test, y_ucz, y_test):
    r = sklearn.linear_model.LinearRegression()
    r.fit(X_ucz, y_ucz)
    y_ucz_pred = r.predict(X_ucz)
    y_test_pred = r.predict(X_test)
    r2 = sklearn.metrics.r2_score
    mse = sklearn.metrics.mean_squared_error
    mae = sklearn.metrics.mean_absolute_error
    return {
        "r_score_u": r2(y_ucz, y_ucz_pred),
        "r_score_t": r2(y_test, y_test_pred),
        "MSE_u": mse(y_ucz, y_ucz_pred),
        "MSE_t": mse(y_test, y_test_pred),
        "MAE_u": mae(y_ucz, y_ucz_pred),
        "MAE_t": mae(y_test, y_test_pred)
    }
````

````python
params = ["Reg. liniowa"]
res = [fit_regression(X_ucz, X_test, y_ucz, y_test)]
pd.DataFrame(res, index=params)
````

### Przybliżę teraz wartość po przecinku zmiennej *rating*, aby zobaczyć jaki będzie miało to wpływ na zdolność modelu do przewidywania wartości

````python
choco.rating = np.int64(choco.rating)
choco.info()
````

````python
choco.describe()
````

````python
X = choco.drop("rating", axis=1)
y = choco["rating"]
X_ucz, X_test, y_ucz, y_test = sklearn.model_selection.train_test_split(
    X, y, test_size=0.2, random_state=9063
)
params.append("Reg. liniowa, int rating")
res.append(fit_regression(X_ucz, X_test, y_ucz, y_test))
pd.DataFrame(res, index=params)
````

## Zaokrąglenie wartości zmiennych nie zdaje się mieć wpływu na działanie modelu. 

## Aby usprawnić model możemy spróbować usunąć zmienną o największej korelacji, której zdecydowałem się wcześniej nie usuwać - zawartość słodzika.
